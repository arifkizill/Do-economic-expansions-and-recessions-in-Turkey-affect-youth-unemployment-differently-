# Economic Cycles and Youth Unemployment in Türkiye

## 1. Import Libraries

```python
import pandas as pd
```

## 2. Load Data

```python
gdp = pd.read_csv("../data/gdp_growth.csv", skiprows=4)
unemployment = pd.read_csv("../data/total_unemployment.csv", skiprows=4)
youth = pd.read_csv("../data/youth_unemployment.csv", skiprows=4)
```

## 3. Filter Türkiye Data

```python
gdp = gdp[gdp["Country Name"] == "Turkiye"]
unemployment = unemployment[unemployment["Country Name"] == "Turkiye"]
youth = youth[youth["Country Name"] == "Turkiye"]
```

## 4. Select Relevant Years

```python
years = [str(year) for year in range(1991, 2025)]

gdp = gdp[years]
unemployment = unemployment[years]
youth = youth[years]
```

## 5. Transpose and Prepare Data

```python
gdp = gdp.T
unemployment = unemployment.T
youth = youth.T

gdp.columns = ["GDP_Growth"]
unemployment.columns = ["Total_Unemployment"]
youth.columns = ["Youth_Unemployment"]

gdp.index.name = "Year"
unemployment.index.name = "Year"
youth.index.name = "Year"
```

## 6. Merge Data

```python
data = pd.concat([gdp, unemployment, youth], axis=1)

data = data.apply(pd.to_numeric, errors="coerce")

data.head()
```

## 10. Regression Analysis

```python
import statsmodels.api as sm

X = data[["GDP_Growth"]]

X = sm.add_constant(X)

y = data["Youth_Unemployment"]

model = sm.OLS(y, X).fit()

print(model.summary())
```

## 11. Scatter Plot and Regression Line

```python
plt.figure(figsize=(10,6))

plt.scatter(
    data["GDP_Growth"],
    data["Youth_Unemployment"],
    s=70
)

plt.plot(
    data["GDP_Growth"],
    model.predict(X),
    linewidth=2
)

plt.title("Relationship Between GDP Growth and Youth Unemployment", fontsize=15)

plt.xlabel("GDP Growth (%)")
plt.ylabel("Youth Unemployment (%)")

plt.grid(alpha=0.3)

plt.tight_layout()

plt.show()
```

## 12. Expansion and Recession Analysis

```python
data["Period"] = data["GDP_Growth"].apply(
    lambda x: "Expansion" if x > 0 else "Recession"
)

data.groupby("Period")["Youth_Unemployment"].mean()
```

## 13. Average Youth Unemployment During Expansions and Recessions

```python
avg_data = data.groupby("Period")["Youth_Unemployment"].mean()

plt.figure(figsize=(7,5))

avg_data.plot(kind="bar")

plt.title("Average Youth Unemployment by Economic Period")
plt.xlabel("Economic Period")
plt.ylabel("Youth Unemployment (%)")

plt.xticks(rotation=0)

plt.grid(alpha=0.3)

plt.tight_layout()

plt.show()
```

## 14. Conclusion

The results of this project show that economic expansions and recessions affect youth unemployment in Türkiye. However, the regression results suggest that GDP growth alone does not strongly explain changes in youth unemployment. This may indicate that other economic and structural factors also play an important role in the youth labor market.
